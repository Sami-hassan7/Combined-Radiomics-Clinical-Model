```python
import nibabel as nib
import matplotlib.pyplot as plt

# Load the NIfTI files
original_file = "/Intern Programming/data/UMCGOPC_MRI_nii_zscoreinHNmask/UMCG-0020715_t1r.nii.gz"
contour_file = "/Intern Programming/data/t1t2_gtv/UMCG-0020715_gtvpttot1r.nii.gz"

original_img = nib.load(original_file)
contour_img = nib.load(contour_file)

# Get the image data from the NIfTI files
original_data = original_img.get_fdata()
contour_data = contour_img.get_fdata()

# Display a slice of the original image and overlay the contour mask
slice_index = original_data.shape[2] // 2  # Adjust this index as needed
slice_original = original_data[:, :, slice_index]
slice_contour = contour_data[:, :, slice_index]

# Plot the original image
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.imshow(slice_original, cmap='gray')
plt.title('MRI Studies',fontsize=32)
plt.axis('off')
plt.tick_params(axis='both', which='major', labelsize=14)

# Plot the contour mask overlaying the original image
plt.subplot(1, 2, 2)
plt.imshow(slice_original, cmap='gray')
plt.imshow(slice_contour, alpha=0.5, cmap='jet')  # Overlay contour mask with transparency
plt.title('Contour Mask Overlay',fontsize=16)
plt.axis('off')
plt.tick_params(axis='both', which='major', labelsize=14)

plt.show()
```

```python
import nibabel as nib
import matplotlib.pyplot as plt

# Load the NIfTI files
original_file = "/Intern Programming/data/UMCGOPC_MRI_nii_zscoreinHNmask/UMCG-0020715_t1r.nii.gz"
contour_file = "/Intern Programming/data/output/t1t2gtv_1mm/UMCG-0020715_gtvpttot1r.nii.gz"

original_img = nib.load(original_file)
contour_img = nib.load(contour_file)

# Get the image data from the NIfTI files
original_data = original_img.get_fdata()
contour_data = contour_img.get_fdata()

# Display a slice of the original image and overlay the contour mask
slice_index = original_data.shape[2] // 2  # Adjust this index as needed
slice_original = original_data[:, :, slice_index]
slice_contour = contour_data[:, :, slice_index]

# Plot the original image and contour mask separately
plt.figure(figsize=(18, 9))

# Plot the original image
plt.subplot(1, 3, 1)
plt.imshow(slice_original, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Plot the contour mask overlaying the original image
plt.subplot(1, 3, 2)
plt.imshow(slice_original, cmap='gray')
plt.imshow(slice_contour, alpha=0.5, cmap='jet')  # Overlay contour mask with transparency
plt.title('Contour Mask Overlay 5mm')
plt.axis('off')

# Plot the contour mask separately
plt.subplot(1, 3, 3)
plt.imshow(slice_contour, cmap='jet')  # Show contour mask only
plt.title('Contour Mask')
plt.axis('off')

plt.show()
```

```python



import matplotlib.pyplot as plt

# Years and corresponding number of publications for each modality
years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
publications_ct = [5, 10, 15, 25, 50, 100, 200, 300, 400, 500, 800]
publications_pet = [2, 5, 10, 20, 40, 80, 150, 200, 250, 300, 500]
publications_mri = [1, 3, 5, 10, 20, 40, 80, 120, 150, 180, 250]
publications_ultrasound = [1, 2, 3, 5, 10, 20, 30, 40, 50, 60, 80]

# Plot
plt.figure(figsize=(10, 6))

# Plot CT data
plt.plot(years, publications_ct, marker='o', linestyle='-', color='blue', label='CT')

# Plot PET data
plt.plot(years, publications_pet, marker='o', linestyle='-', color='green', label='PET')

# Plot MRI data
plt.plot(years, publications_mri, marker='o', linestyle='-', color='red', label='MRI')

# Plot Ultrasound data
plt.plot(years, publications_ultrasound, marker='o', linestyle='-', color='orange', label='Ultrasound')

plt.title('Growth of Radiomics Research by Modality from 2012 till now', fontsize=16 )
plt.xlabel('Year')
plt.ylabel('Number of Publications')
plt.grid(True)
plt.xticks(years)
plt.legend()
plt.tick_params(axis='both', which='major', labelsize=18)
plt.tight_layout()
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
from mpl_toolkits.mplot3d.art3d import Poly3DCollection
import skimage.measure

# Example data for tumor intensity, shape, and texture
tumor_intensity = np.random.normal(loc=100, scale=15, size=1000)
voxel_data = (np.indices((10, 10, 10))[0] - 5)**2 + (np.indices((10, 10, 10))[1] - 5)**2 + (np.indices((10, 10, 10))[2] - 5)**2 < 25
texture_data_a = np.random.randint(0, 10, size=(5, 5))
texture_data_b = np.random.randint(0, 10, size=(5, 5))
radiomic_features = np.random.rand(10, 10)

fig = plt.figure(figsize=(16, 8))

# Tumor Intensity
ax1 = fig.add_subplot(231)
ax1.hist(tumor_intensity, bins=50, color='red')
ax1.set_title('Intensity Features')
ax1.set_xlabel('Intensity Value')
ax1.set_ylabel('Frequency')

ax2 = fig.add_subplot(232)
ax2.hist(tumor_intensity, bins=50, color='blue')
ax2.set_title('Intensity B')
ax2.set_xlabel('Intensity Value')
ax2.set_ylabel('Frequency')

# Tumor Shape
ax3 = fig.add_subplot(233, projection='3d')
verts, faces, _, _ = skimage.measure.marching_cubes(voxel_data, level=0.5)
mesh1 = Poly3DCollection(verts[faces], alpha=0.7, edgecolor='k')
ax3.add_collection3d(mesh1)
ax3.set_xlim(0, voxel_data.shape[0])
ax3.set_ylim(0, voxel_data.shape[1])
ax3.set_zlim(0, voxel_data.shape[2])
ax3.set_title('Shape Features')

ax4 = fig.add_subplot(234, projection='3d')
mesh2 = Poly3DCollection(verts[faces], alpha=0.7, edgecolor='k')
ax4.add_collection3d(mesh2)
ax4.set_xlim(0, voxel_data.shape[0])
ax4.set_ylim(0, voxel_data.shape[1])
ax4.set_zlim(0, voxel_data.shape[2])
ax4.set_title('Tumor Shape B')

# Tumor Texture
ax5 = fig.add_subplot(235)
sns.heatmap(texture_data_a, annot=True, fmt='d', cmap='viridis', ax=ax5)
ax5.set_title('Texture Features')

ax6 = fig.add_subplot(236)
sns.heatmap(texture_data_b, annot=True, fmt='d', cmap='viridis', ax=ax6)
ax6.set_title('Tumor Texture B')

plt.tight_layout()
plt.show()

# Radiomic Features
plt.figure(figsize=(6, 6))
sns.heatmap(radiomic_features, cmap='coolwarm', annot=False)
plt.title('Radiomic Features')
plt.show()

# Clinical feature
plt.figure(figsize=(6, 6))
sns.heatmap(radiomic_features, cmap='cividis', annot=False)
plt.title('Clinical data')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from skimage import measure
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

# Generating synthetic data for demonstration
tumor_intensity = np.random.normal(loc=100, scale=15, size=1000)
voxel_data = (np.indices((10, 10, 10))[0] - 5)**2 + (np.indices((10, 10, 10))[1] - 5)**2 + (np.indices((10, 10, 10))[2] - 5)**2 < 25
texture_data_a = np.random.randint(0, 10, size=(5, 5))
texture_data_b = np.random.randint(0, 10, size=(5, 5))
radiomic_features = np.random.rand(10, 10)

fig = plt.figure(figsize=(16, 8))

# Tumor Intensity
ax1 = fig.add_subplot(231)
ax1.hist(tumor_intensity, bins=50, color='red')
ax1.set_title('Intensity Features', fontsize=16)
ax1.set_xlabel('Intensity Value', fontsize=16)
ax1.set_ylabel('Frequency', fontsize=16)
ax1.tick_params(axis='both', which='major', labelsize=14)

ax2 = fig.add_subplot(232)
ax2.hist(tumor_intensity, bins=50, color='blue')
ax2.set_title('Tumor Intensity B', fontsize=14)
ax2.set_xlabel('Intensity Value', fontsize=12)
ax2.set_ylabel('Frequency', fontsize=12)
ax2.tick_params(axis='both', which='major', labelsize=10)

# Tumor Shape
ax3 = fig.add_subplot(233, projection='3d')
verts, faces, _, _ = measure.marching_cubes(voxel_data, level=0.5)
mesh1 = Poly3DCollection(verts[faces], alpha=0.7, edgecolor='k', facecolor=plt.cm.viridis(0.5))
ax3.add_collection3d(mesh1)
ax3.set_xlim(0, voxel_data.shape[0])
ax3.set_ylim(0, voxel_data.shape[1])
ax3.set_zlim(0, voxel_data.shape[2])
ax3.set_title('Shape Features', fontsize=16)
ax3.tick_params(axis='both', which='major', labelsize=14)

ax4 = fig.add_subplot(234, projection='3d')
mesh2 = Poly3DCollection(verts[faces], alpha=0.7, edgecolor='k', facecolor=plt.cm.plasma(0.5))
ax4.add_collection3d(mesh2)
ax4.set_xlim(0, voxel_data.shape[0])
ax4.set_ylim(0, voxel_data.shape[1])
ax4.set_zlim(0, voxel_data.shape[2])
ax4.set_title('Tumor Shape B', fontsize=16)
ax4.tick_params(axis='both', which='major', labelsize=14)

# Tumor Texture
ax5 = fig.add_subplot(235)
sns.heatmap(texture_data_a, annot=True, fmt='d', cmap='viridis', ax=ax5)
ax5.set_title('Texture Features', fontsize=16)
ax5.tick_params(axis='both', which='major', labelsize=14)

ax6 = fig.add_subplot(236)
sns.heatmap(texture_data_b, annot=True, fmt='d', cmap='plasma', ax=ax6)
ax6.set_title('Tumor Texture B', fontsize=16)
ax6.tick_params(axis='both', which='major', labelsize=14)

plt.tight_layout()
plt.show()

# Radiomic Features
plt.figure(figsize=(4, 3))
sns.heatmap(radiomic_features, cmap='coolwarm', annot=False)
plt.title('Radiomics Features', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()

plt.figure(figsize=(4, 3))
sns.heatmap(radiomic_features, cmap='cividis', annot=False)
plt.title('Clinical data', fontsize=16)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()
```
